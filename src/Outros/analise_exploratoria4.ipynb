```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv("Walmart_Sales.csv")

# Converter a coluna 'Date' para formato de data
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")

# Converter a temperatura de Fahrenheit para Celsius
df['Temperature'] = (df['Temperature'] - 32) * 5/9

# Criar gráfico de dispersão entre Temperatura e Vendas Semanais
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Temperature", y="Weekly_Sales", alpha=0.5)
plt.title("Relação entre Temperatura e Vendas Semanais")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Vendas Semanais ($)")
plt.grid(True)
plt.show()

# Criar matriz de correlação
sns.set(style="whitegrid")
correlation_matrix = df.drop(['Store', 'Date'], axis=1).corr()

# Gerar um mapa de calor da matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlação')
plt.show()

# Criar coluna de trimestre
df['Quarter'] = df['Date'].dt.to_period('Q')

# Calcular vendas totais por trimestre
vendas_por_trimestre = df.groupby('Quarter')['Weekly_Sales'].sum().reset_index()

# Criar gráfico de evolução de vendas por trimestre
plt.figure(figsize=(10, 6))
sns.lineplot(data=vendas_por_trimestre, x=vendas_por_trimestre['Quarter'].astype(str), y='Weekly_Sales', marker='o')
plt.title('Evolução das Vendas por Trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Vendas Totais ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv("Walmart_Sales.csv")

# Converter a coluna 'Date' para formato de data
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")

# Converter a temperatura de Fahrenheit para Celsius
df['Temperature'] = (df['Temperature'] - 32) * 5/9

# Criar gráfico de dispersão entre Temperatura e Vendas Semanais
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Temperature", y="Weekly_Sales", alpha=0.5)
plt.title("Relação entre Temperatura e Vendas Semanais")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Vendas Semanais ($)")
plt.grid(True)
plt.show()

# Criar matriz de correlação
sns.set(style="whitegrid")
correlation_matrix = df.drop(['Store', 'Date'], axis=1).corr()

# Gerar um mapa de calor da matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlação')
plt.show()

# Criar coluna de trimestre
df['Quarter'] = df['Date'].dt.to_period('Q')

# Calcular vendas totais por trimestre
vendas_por_trimestre = df.groupby('Quarter')['Weekly_Sales'].sum().reset_index()

# Determinar se os valores estão em milhões ou bilhões
vendas_por_trimestre['Weekly_Sales'] = vendas_por_trimestre['Weekly_Sales'] / 1e6  # Convertendo para milhões

# Criar gráfico de evolução de vendas por trimestre
plt.figure(figsize=(10, 6))
sns.lineplot(data=vendas_por_trimestre, x=vendas_por_trimestre['Quarter'].astype(str), y='Weekly_Sales', marker='o')
plt.title('Evolução das Vendas por Trimestre (em Milhões)')
plt.xlabel('Trimestre')
plt.ylabel('Vendas Totais (Milhões $)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Criar gráfico de volume de vendas por trimestre
plt.figure(figsize=(10, 6))
sns.barplot(data=vendas_por_trimestre, x=vendas_por_trimestre['Quarter'].astype(str), y='Weekly_Sales', palette='Blues')
plt.title('Volume de Vendas por Trimestre (em Milhões)')
plt.xlabel('Trimestre')
plt.ylabel('Vendas Totais (Milhões $)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
