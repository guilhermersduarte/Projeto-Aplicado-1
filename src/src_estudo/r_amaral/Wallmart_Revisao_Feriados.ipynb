```r
dfreport = read.csv("./boom_bust_by_store.csv")
options(repr.plot.width=8, repr.plot.height=8)
# Load the data (assuming it's already in a data frame called 'data')
# For this script, we assume the data is already loaded with the given column names

# Set up a 2x2 plotting grid
par(mfrow=c(2,2), mar=c(4,4,2,1))

# First quadrant: Boxplot of semanas_pre
boxplot(dfreport$semanas_pre, main="Semanas de Boom", ylab="Weeks", col="lightblue", border="black")

# Second quadrant: Boxplot of boom_above_neutral_percent
boxplot(dfreport$boom_above_neutral_percent, main="Aumento de Vendas em Boom", ylab="Percentage", col="lightgreen", border="black")

# Third quadrant: Boxplot of semanas_pos
boxplot(dfreport$semanas_pos, main="Semanas de Bust", ylab="Weeks", col="lightblue", border="black")

# Fourth quadrant: Boxplot of bust_below_neutral_percent
boxplot(dfreport$bust_below_neutral_percent, main="Redução de Vendas em Bust", ylab="Percentage", col="lightcoral", border="black")

# Reset plotting parameters
par(mfrow=c(1,1))
```

```r
# Set up a 2x2 plotting grid
options(repr.plot.width=8, repr.plot.height=8)
par(mfrow=c(2,2), mar=c(4,4,2,1))

# First quadrant: Histogram of semanas_pre

hist(dfreport$semanas_pre, 
     main = "Semanas de Boom",
     xlab = "Weeks",
     ylab = "Frequency",
     col = "lightblue",
     border = "white",
     breaks=4)

# Second quadrant: Histogram of boom_above_neutral_percent
hist(dfreport$boom_above_neutral_percent, 
     main = "Aumento de Vendas em Boom",
     xlab = "Percentage",
     ylab = "Frequency",
     col = "lightgreen",
     border = "white",
     breaks = 6)

# Third quadrant: Histogram of semanas_pos
hist(dfreport$semanas_pos, 
     main = "Semanas de Bust",
     xlab = "Weeks",
     ylab = "Frequency",
     col = "lightblue",
     border = "white",
     breaks = 4)

# Fourth quadrant: Histogram of bust_below_neutral_percent
hist(dfreport$bust_below_neutral_percent, 
     main = "Redução de Vendas em Bust",
     xlab = "Percentage",
     ylab = "Frequency",
     col = "lightcoral",
     border = "white",
     breaks = 10)

# Reset plotting parameters
par(mfrow=c(1,1))


#hist(data, 
#     main = "Histogram of Random Normal Data",
#     xlab = "Value",
#     ylab = "Frequency",
#     col = "lightblue",
#     border = "white",
#     breaks = 30)
```
