A inflação no formato CPI é a variação a partir de uma data inicial predeterminada. Para que possamos avaliar o impacto da inflação semanal nas vendas é nessessário calcular o variação semanal do valor absoluto. O cálculo deve ser feito por loja.

```r
#Calcular, em percentagem (*100) a variação semanal do CPI para cada loja.
for (nstore in 1:45) {


    df[df$Store == nstore, "Weekly_CPI"] = c(NA, diff(df[df$Store == nstore, "CPI"]) / df[df$Store == nstore, "CPI"][-length(df[df$Store == nstore, "CPI"])]) * 100
    
}
```

```r
n <- 20  # Imprimir 20 linhas aleatórias para conferir o cáculo
random_indices <- sample(nrow(df), n)
random_rows <- df[random_indices, ]
print(random_rows)
```
