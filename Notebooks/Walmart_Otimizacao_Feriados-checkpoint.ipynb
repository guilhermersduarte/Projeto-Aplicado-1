```r
suppressPackageStartupMessages(library(dplyr))
library(dplyr)
df = read.csv("./Walmart_Sales_Updated.csv")
```

```r
#Conferir se já não foi alterado e converter coluna 'Date pata formato Date
if ( class(df$Date) != "Date" ) {
  df$Date = as.Date(df$Date, format = "%Y-%m-%d")
}
dfreport = data.frame(
  loja = integer(),
  semanas_pre = integer(),
  semanas_pos = integer(),
  delta = numeric(),
  delta_percent = numeric(),
  boom_above_neutral_percent = numeric(),
  bust_below_neutral_percent = numeric()
)
```

```r
head(df)
head(dfreport)
```

Estudando o efeito de feriados

```r
for (nstore in 0:45){
    
    if (nstore > 0){
        dfplot = df[(df$Store == nstore),]
    } else {
      dfplot = df %>%
      select(Date, Weekly_Sales, Holiday_Flag) %>%
      group_by(Date) %>%
      summarise(
        Weekly_Sales = sum(Weekly_Sales, na.rm = TRUE),
        Holiday_Flag = mean(Holiday_Flag, na.rm = TRUE)
      )
    }

    sales_mean = mean(dfplot$Weekly_Sales)
                      
    final_weeks_pre = 0
    final_weeks_post = 0
    final_delta = 0
    final_delta_percent = 0
    final_boom_above_neutral_percent = 0
    final_bust_below_neutral_percent = 0
    
    
    for (weeks_pre in 1:8) {
        
        for  (weeks_post in 1:8){
            
            condition = dfplot$Holiday_Flag == 1
            holiday_dates = dfplot$Date[which(condition)]
            nholidays = length(holiday_dates)
            
            
            
            ##Pre Holiday
            dfplot$Pre_Holiday = 0
            for (i in 1:nholidays){
                
                match_date = dfplot$Date[which(condition)[i]]
            
                dfplot$Pre_Holiday = ifelse(dfplot$Date <= match_date & dfplot$Date >= (match_date - weeks_pre*7),1,dfplot$Pre_Holiday)
                
                }
            
            #Post Holiday
            dfplot$Post_Holiday = 0
            for (i in 1:nholidays){
                
                match_date = dfplot$Date[which(condition)[i]]
            
                dfplot$Post_Holiday = ifelse(dfplot$Date > match_date & dfplot$Date <= (match_date + weeks_post*7),1,dfplot$Post_Holiday)
                
                }
            
            dfplot$Holiday_Effect = 'Neutral'
            
            dfplot$Holiday_Effect = ifelse(dfplot$Post_Holiday == 1,'Bust', dfplot$Holiday_Effect)
            dfplot$Holiday_Effect = ifelse(dfplot$Pre_Holiday == 1,'Boom', dfplot$Holiday_Effect)
            
            #Calcula médias de venda para a combinação LOJA/PRE/PÓS
            means <- aggregate(Weekly_Sales ~ Holiday_Effect, dfplot, mean)
            #print(means)
            
            #cáculo do delta Boom/Bust
            delta = means$Weekly_Sales[means$Holiday_Effect  == "Boom"] - means$Weekly_Sales[means$Holiday_Effect  == "Bust"]
            delta_percent = (delta/ means$Weekly_Sales[means$Holiday_Effect  == "Neutral"])*100
            boom_above_neutral_percent = ((means$Weekly_Sales[means$Holiday_Effect  == "Boom"] - means$Weekly_Sales[means$Holiday_Effect  == "Neutral"])/means$Weekly_Sales[means$Holiday_Effect  == "Neutral"])*100
            bust_below_neutral_percent = ((means$Weekly_Sales[means$Holiday_Effect  == "Neutral"] - means$Weekly_Sales[means$Holiday_Effect  == "Bust"])/means$Weekly_Sales[means$Holiday_Effect  == "Neutral"])*100
    
            if (delta_percent > final_delta_percent) {
    
                final_weeks_pre = weeks_pre
                final_weeks_post = weeks_post
                final_delta = delta
                final_delta_percent = delta_percent
                final_boom_above_neutral_percent = boom_above_neutral_percent
                final_bust_below_neutral_percent = bust_below_neutral_percent
    
            }
    
        }
    
    }

    dfreport = rbind(dfreport, data.frame(
    loja = nstore,
    semanas_pre = final_weeks_pre,
    semanas_pos = final_weeks_post,
    delta = final_delta,
    delta_percent = final_delta_percent,
    boom_above_neutral_percent = final_boom_above_neutral_percent,
    bust_below_neutral_percent = final_bust_below_neutral_percent
        
    ))
    #print(paste0("Loja: ", nstore))
    #print(paste0("Semanas Boom: ", final_weeks_pre))
    #print(paste0("Semanas Bust: ", final_weeks_post))
    #print(paste0("Variação média $ Boom/Bust : $", format(final_delta, big.mark = ",", nsmall = 2)))
    #print(paste0("Variação média % Boom/Bust : ", format(final_delta_percent,  ,nsmall = 2), "%"))
}
options(width = 1000)
print(dfreport)
```

Criar boxplot com três grupos, pre (boom), pós (bust) e indiferente (neutral).

Fica clara a influência da aproximação do feriado no aumento de vandas e depois a ressaca de vendas depois do feriado.
