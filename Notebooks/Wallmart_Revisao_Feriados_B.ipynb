```r
dfreport = read.csv("./boom_bust_by_store.csv")

options(repr.plot.width=9, repr.plot.height=6)

par(mfrow=c(2,3),        
    mar=c(4,4,2,1),      
    oma=c(2,3,2,3)) 


boxplot(dfreport$semanas_pre, main="Semanas de Boom", ylab="Weeks", col="lightblue", border="black")

boxplot(dfreport$boom_above_neutral_percent, main="Aumento de Vendas em Boom", ylab="Percentage", col="lightgreen", border="black")

hist(dfreport$boom_above_neutral_percent, 
     main = "Aumento de Vendas em Boom",
     xlab = "Percentage",
     ylab = "Frequency",
     col = "lightgreen",
     border = "white",
     breaks = 6)

boxplot(dfreport$semanas_pos, main="Semanas de Bust", ylab="Weeks", col="lightblue", border="black")

boxplot(dfreport$bust_below_neutral_percent, main="Redução de Vendas em Bust", ylab="Percentage", col="lightcoral", border="black")

hist(dfreport$bust_below_neutral_percent, 
     main = "Redução de Vendas em Bust",
     xlab = "Percentage",
     ylab = "Frequency",
     col = "lightcoral",
     border = "white",
     breaks = 15)


par(mfrow=c(1,1))
#dev.off()
```
